```python
N = 10**4
theta = 1
```

```python
sample = sps.expon.rvs(size = N, scale= 1 / theta)
```

```python
n = np.arange(1, N + 1)
```

### –ì—Ä–∞—Ñ–∏–∫ –º–æ–¥—É–ª—è —Ä–∞–∑–Ω–æ—Å—Ç–∏ –æ—Ü–µ–Ω–∫–∏ –∏ –∏—Å—Ç–∏–Ω–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è ùúÉ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç ùëõ (ùëõ $\leq$ N)

```python
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'pink', 'gold',
          'navy', 'springgreen', 'chocolate', 'aqua' ]
plt.figure(figsize=(15, 8))
plt.xlim((2000, N))  # —Ä–∞–∑–º–µ—Ä—ã –≥—Ä–∞—Ñ–∏–∫–∞ –ø–æ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–π –æ—Å–∏ 
plt.ylim((0, 0.25))  # —Ä–∞–∑–º–µ—Ä—ã –≥—Ä–∞—Ñ–∏–∫–∞ –ø–æ –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–æ–π –æ—Å–∏ 
for k in range(12):
    plt.plot(n, abs(estimate(sample,k+1) - theta), color= colors[k], 
             linewidth=2.5, label= k + 1) 
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel('N')
plt.ylabel('Estimate of the difference module')
plt.grid()  # —Å–µ—Ç–∫–∞
plt.title(r'$\theta=$' + str(theta))
```
