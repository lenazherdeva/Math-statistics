```python
import pandas as pd
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
%pylab inline
```

```python
N = 10**4
```

## Выборка при $\theta=$ 5

```python
theta = 5
```

```python
sample = sps.uniform.rvs(size = N, loc=0, scale= theta)
n = np.arange(1, N+1)
```

```python
means = sample.cumsum() / n  # массив из частичных средних
```

```python
first = 2 * means
```

```python
curr_min = sample[0]
curr_max = sample[0]
second = numpy.zeros(N)
third = numpy.zeros(N)
fourth = numpy.zeros(N)
fifth = numpy.zeros(N)
for i in range(N) :
    curr_min = min(curr_min, sample[i])
    curr_max = max(curr_max, sample[i])
    second[i] = means[i] + curr_max/2
    m = i + 1
    third[i] = (m+1) * curr_min
    fourth[i] = curr_max + curr_min
    fifth[i] = ((m+1) / m ) * curr_max
```

# Графики для  $\theta=$ 5

```python
plt.figure(figsize=(15, 8))
plt.plot(n, abs(first - theta), color='red', linewidth=2.5, label= '2$\overline{X}$') 
plt.plot(n, abs(second - theta), color='blue', linewidth=2.5, label= '$\overline{X}+X_{(n)}/2$')
plt.plot(n, abs(third - theta), color='yellow', linewidth=2.5, label= '$(n+1)X_{(1)}$')
plt.plot(n, abs(fourth - theta), color='purple', linewidth=2.5, label= '$X_{(1)}+X_{(n)}$')
plt.plot(n, abs(fifth - theta), color='green', linewidth=2.5, label= r'$\frac{n+1}{n}X_{(n)}$')
plt.legend()
plt.xlim((1, N))  # размеры графика по горизонтальной оси 
plt.ylim((0, 6))  # размеры графика по вертикальной оси 
plt.xlabel('N')
plt.ylabel('Estimate of the difference module')
plt.title(r'$\theta=$' + str(theta))
plt.grid()  # сетка
```

### Графики без $(n+1)X_{(1)}$

```python
plt.figure(figsize=(15, 8))
plt.plot(n, abs(first - theta), color='red', linewidth=2.5, label= '2$\overline{X}$') 
plt.plot(n, abs(second - theta), color='blue', linewidth=2.5, label= '$\overline{X}+X_{(n)}/2$')
plt.plot(n, abs(fourth - theta), color='purple', linewidth=2.5, label= '$X_{(1)}+X_{(n)}$')
plt.plot(n, abs(fifth - theta), color='green', linewidth=2.5, label= r'$\frac{n+1}{n}X_{(n)}$')
plt.legend()
plt.xlim((1, N))  # размеры графика по горизонтальной оси 
plt.ylim((0, 1))  # размеры графика по вертикальной оси 
plt.xlabel('N')
plt.ylabel('Estimate of the difference module')
plt.title(r'$\theta=$' + str(theta))
plt.grid()  # сетка
```
